# Airbnb Dataset Profiling Summary

## Dataset Overview

The dataset contains three relational tables representing the Airbnb business domain.

| Table | Description | Rows |
|------|-------------|------|
| Hosts | Information about property hosts | 200 |
| Listings | Details of properties listed on Airbnb | 500 |
| Bookings | Reservation transactions for listings | 5000 |

---

## Dataset Relationships

Hosts → Listings → Bookings

- A **host** can manage multiple listings  
- A **listing** can receive multiple bookings  
- Each **booking belongs to one listing**

Relationship keys:

hosts.host_id → listings.host_id  
listings.listing_id → bookings.listing_id

---

## Data Quality Summary

| Check | Result |
|------|-------|
| Missing Values | None detected |
| Duplicate Records | None detected |
| Primary Key Integrity | All primary keys unique |
| Referential Integrity | Valid relationships across tables |
| Numerical Ranges | Within expected limits |
| Dataset Completeness | High quality dataset |

Overall, the dataset demonstrates **strong structural integrity** and is suitable for building a data warehouse pipeline.

---

## Key Observations

### Data Completeness
- No missing values detected across any dataset

### Duplicate Checks
- No duplicate rows detected
- All primary keys (`host_id`, `listing_id`, `booking_id`) are unique

### Referential Integrity
- All listings reference valid hosts
- All bookings reference valid listings
- No orphan records found

### Data Types
Most numerical fields are correctly stored as integers.

However, the following date fields are stored as strings:

- host_since  
- booking_date  
- created_at  

These will be converted to **timestamp formats in the Silver layer**.

### Numerical Distributions

Observed ranges appear realistic:

- Host response rate: **70 – 100**
- Price per night: **50 – 300**
- Nights booked: **1 – 14**
- Booking amount: **50 – ~4172**

No negative or unrealistic values were detected.

---

## Planned Transformations (Silver Layer)

Based on profiling results, the following transformations will be implemented in **dbt Silver models**.

### 1️⃣ Date Standardization
Convert string fields to timestamps:

- host_since
- booking_date
- created_at

### 2️⃣ Data Type Enforcement
Apply explicit type casting:

- price_per_night → numeric
- response_rate → integer

### 3️⃣ Business Rule Validation
Ensure valid values:

- nights_booked > 0
- price_per_night > 0
- response_rate ≤ 100

### 4️⃣ Feature Engineering
Create analytical metrics:

total_revenue = booking_amount + cleaning_fee + service_fee

### 5️⃣ Data Integrity Preservation
Maintain valid relationships:

hosts.host_id → listings.host_id  
listings.listing_id → bookings.listing_id
